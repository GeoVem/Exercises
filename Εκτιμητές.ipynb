```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
from scipy.stats import norm
```

**Gaussian Distribution** $$ f(x; \mu, \sigma)=\displaystyle \frac{1}{\sqrt{2\pi{}\sigma^2}}exp\left(-\frac{(x-\mu)^2}{\sigma^2}\right) $$

```python
mu = 1
sigma = 2
mean, var = norm.stats(loc=mu, scale=sigma, moments='mv')
```

```python
r = norm.rvs( mu, sigma , size = 100000 )
```

```python
r
```

Τυχαιο δείγμα 5 μετρήσεων Χ:


```python
x1 = np.random.choice(r, 5)
```

```python
x1
```

Πρώτος εκτιμητής $Μ = \displaystyle \frac{1}{n}\sum\limits^n_{i=1}(X_i) $ :

```python
M1 = 0 
j = 0
```

```python
while (j < 5):
    M1 = (M1 + x1[j])
    j = j + 1 
    print(M1)
else:
    M1 = (1 / 5 ) * M1
    print(M1)
```

```python
M1
```

Τα παραπάνω ήταν ένα τεστ για να δούμε αν δουλέυει, τώρα απο κάτω θα κανουμε κατι παρόμοιο αλλα για $10^5$ μετρήσεις:
Θα εξετάσουμε ταυτόχρονα και αλλους 2 εκτιμητές: $\displaystyle S^2 = \frac{1}{n}\sum\limits^n_{i=1}(X_i-M)^2, S^{*2}=\frac{1}{n-1}\sum\limits^n_{i=1}(X_i-M)^2$ που έχουν να κάνουν με την διακύμανση.

```python
m_empty = np.array([])
m = np.array([])
s_empty = np.array([])
s = np.array([])
ss_empty = np.array([])
ss = np.array([])
i = 0
```

```python
while ( i <= 100000):    
    x = np.random.choice( r , 5 )
    M = 0
    j = 0
    while (j < 5):
        M = (M + x[j])
        j = j + 1
    else:
        M = (1 / 5) * M
        t = 0 
        S2 = 0
        while (t < 5):
            S2 = S2 + (x[t]-M)**2
            t = t +1
        else:
            S3 = S2 / 4
            S2 = S2 / 5 
            
    
    a1 = np.append(a_empty, M)
    m = np.append(m, a1)
    a2 = np.append(s_empty, S2)
    s = np.append(s, a2)
    a3 = np.append(ss_empty, S3)
    ss = np.append(ss, a3)
    i = i + 1
```

```python
m
```

```python
s
```

```python
ss
```

```python
plt.figure(figsize=(15,10))
plt.style.use('seaborn-whitegrid')
plt.hist(m, bins=100, density=True, linewidth = 0.5, ec='black', stacked=True, range=(-4, 4), color=('gray'))
plt.xlabel('M', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('M evaluator : $10^5$ Experiments, 5 samples')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.style.use('seaborn-whitegrid')
plt.hist(s, bins=100, density=True, linewidth = 0.5, ec='black', stacked=True, color=('gray'), range=(0, 30))
plt.xlabel('$S^2$', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('$S^2$ evaluator : $10^5$ Experiments, 5 samples')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.style.use('seaborn-whitegrid')
plt.hist(ss, bins=100, density=True, linewidth = 0.5, ec='black', stacked=True, color=('gray'), range=(0, 30))
plt.xlabel('$S^{*2}$', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('$S^{*2}$ evaluator : $10^5$ Experiments, 5 samples')
plt.show()
```
