## (i)Gaussian Distribution

$$ f(x; \mu, \sigma)=\displaystyle \frac{1}{\sqrt{2\pi{}\sigma^2}}exp\left(-\frac{(x-\mu)^2}{\sigma^2}\right) $$

```python
from scipy.stats import norm
```

```python
mu = 1
sigma = 2
mean, var = norm.stats(loc=mu, scale=sigma, moments='mv')
```

```python
x = np.linspace(norm.ppf(0.00001, mu, sigma), norm.ppf(0.99999, mu, sigma), 100)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x; μ=1, σ=2)')
plt.plot(x, norm.pdf(x, mu, sigma))
```

```python
r = norm.rvs( mu, sigma , size = 1000 )
```

```python
plt.figure(figsize=(8,8))
plt.hist(r, bins=100, density=False, histtype='step')
plt.xlabel('Normal Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Normal: n=1000, bins=100, σ=2, μ=1')
plt.show()
```

```python
r = norm.rvs( mu, sigma , size = 1000 )
```

```python
plt.figure(figsize=(15,8))
plt.hist(r, bins=50, density=False, histtype='step')
plt.xlabel('Normal Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Normal: n=1000, bins=50, σ=2, μ=1')
plt.show()
```

## (ii) Exponential Distribution

```python
from scipy.stats import expon
```

$$ f(x; \lambda)= \lambda e^{-\lambda x}, \ x \geq 0 $$

```python
lambdaa = 0.5
```

```python
x = np.linspace(expon.ppf(0.00001), expon.ppf(0.9999), 100)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x; λ=0.5)')
plt.plot(x, expon.pdf(x), lambdaa , lw=1, alpha=0.5, label='expon pdf')
```

```python
r1 = expon.rvs(lambdaa , size = 1000)
```

```python
plt.figure(figsize=(15,8))
plt.hist(r1, bins=100, density=False, histtype='step')
plt.xlabel('Exponential Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Exp: n=1000, bins=100, λ=0.5')
plt.show()
```

## (iii)Beta Distribution

$$ f(x;a,b)=\displaystyle \frac{\Gamma(a+b)x^{a-1}(1-x)^{b-1}}{\Gamma(a)\Gamma(b)}, \ 0 \leq x \leq 1$$

```python
from scipy.stats import beta
```

```python
a , b = 2.5 , 3.1
```

```python
x = np.linspace(beta.ppf(0, a, b), beta.ppf(1, a, b), 100)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x; a=2.5 ,b=3.1)')
plt.plot(x, beta.pdf(x, a, b))
```

```python
r2 = beta.rvs(a, b, size = 1000)
```

```python
plt.figure(figsize=(20,8))
plt.hist(r2, bins=100, density=False, histtype='step')
plt.xlabel('Beta Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Beta: n=1000, bins=100, a=2.5, b=3.1')
plt.show()
```

```python
r2 = beta.rvs(a, b, size = 1000)
```

```python
plt.figure(figsize=(20,15))
plt.hist(r2, bins=50, density=False, histtype='step')
plt.xlabel('Beta Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Beta: n=1000, bins=50, a=2.5, b=3.1')
plt.show()
```

```python
r2 = beta.rvs(a, b, size = 10000)
```

```python
plt.figure(figsize=(20,15))
plt.hist(r2, bins=100, density=False, histtype='step')
plt.xlabel('Beta Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.title('Beta: n=10000, bins=100, a=2.5, b=3.1')
plt.show()
```
